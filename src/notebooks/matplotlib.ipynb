### plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib-axes-axes-plot

`plot` 메서드를 통해 라인플롯을 그릴 수  있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(-10, 10, 1000)
y = lambda x: np.sin(x)
ax.plot(x, y(x))
ax.grid()
plt.show()
```

### scatter
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib-axes-axes-hist

`scatter` 메서드를 통해 스캐터플롯을 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(-10, 10, 1000)
y = lambda x: np.sin(x)
noise = np.random.normal(
    loc=0, 
    scale=0.1, 
    size=x.shape
)

ax.scatter(x=x, 
           y=y(x)+noise, 
           s=2.5, 
           color="magenta")
ax.grid()
plt.plot()
```

### bar
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib-axes-axes-bar

`bar` 메서드를 통해 막대그래프를 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.arange(-10, 10)
y = lambda x: np.sin(x)
noise = np.random.normal(
    loc=0, 
    scale=0.1, 
    size=x.shape
)

ax.bar(x=x, 
       height=y(x)+noise, 
       color="darkblue")
ax.grid()
plt.plot()
```

### hist
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib-axes-axes-hist

`hist` 메서드를 통해 히스토그램을 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(-10, 10, 1000)
y = lambda x: np.sin(x)
samples = np.random.normal(
    loc=5,
    scale=1, 
    size=1000
)
ax.hist(samples, bins=25)
plt.plot()
```

### ticker.Locator
https://matplotlib.org/stable/api/ticker_api.html#tick-locating

`plt.ticker`에는 $x$ 혹은 $y$축의 눈금(tick)을 설정할 수 있는 Locator 클래스들이 있습니다. 

```python
from matplotlib import ticker
import matplotlib.pyplot as plt
fig, axes = plt.subplots(
    nrows=4,
    figsize=(6,7),
    subplot_kw={
        "xlim": [-5, 5],
        "ylim": [-0.5, 0.5],
        "yticks": [],
    },
)

for ax in axes:
    ax.spines[["top", "left", "right"]].set_visible(False)


axes[0].xaxis.set_major_locator(
    ticker.NullLocator()
)
axes[0].text(-4.8, -0.3, 
            s= "NullLocator()", 
            fontsize=14)


axes[1].xaxis.set_major_locator(
    ticker.FixedLocator([-5, 0, 5])
)
axes[1].text(-4.8, -0.3, 
        s='FixedLocator([-5, 0, 5])', 
        fontsize=14)

axes[2].xaxis.set_major_locator(
    ticker.MultipleLocator(2.5)
)
axes[2].text(-4.8, -0.3, 
            s='ticker.MultipleLocator(2.5)', 
            fontsize=14)


axes[3].xaxis.set_major_locator(
    ticker.FixedLocator([-5, 0, 5])
)
axes[3].xaxis.set_minor_locator(
    ticker.MultipleLocator(1)
)
axes[3].text(-4.8, -0.3, 
    s="set_minor_locator(ticker.MultipleLocator(1))", 
    fontsize=14)

plt.show()
```

### ticker.Formatter
https://matplotlib.org/stable/api/ticker_api.html#tick-formatting

Locator와 더불어 Formatter 클래스들을 통해 눈금에 대한 레이블을 지정할 수 있습니다.

```python
from matplotlib import ticker
import matplotlib.pyplot as plt
fig, axes = plt.subplots(
    nrows=5,
    figsize=(6,10),
    subplot_kw={
        "xlim": [-5, 5],
        "ylim": [-0.5, 0.5],
        "yticks": [],
    },
)

for ax in axes:
    ax.spines[["top", "left", "right"]].set_visible(False)
    ax.xaxis.set_major_locator(
        ticker.FixedLocator([-5, 0, 5])
    )

axes[0].xaxis.set_major_formatter(
    ticker.NullFormatter()
)
axes[0].text(-4.8, -0.3, 
            s= "NullFormatter()", 
            fontsize=14)


axes[1].xaxis.set_major_formatter(
    ticker.FixedFormatter(["a","b","c"])
)
axes[1].text(-4.8, -0.3, 
        s='FixedFormatter(["a","b","c"])', 
        fontsize=14)


axes[2].xaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f"{x}xyz")
)
axes[2].text(-4.8, -0.3, 
            s='FuncFormatter(lambda x, pos: f"{x}xyz")', 
            fontsize=14)


axes[3].xaxis.set_major_formatter(
    ticker.PercentFormatter(xmax=5)
)
axes[3].text(-4.8, -0.3, 
            s="PercentFormatter(xmax=5)", 
            fontsize=14)


axes[4].xaxis.set_major_formatter(
    ticker.FormatStrFormatter("ab%dcd")
)
axes[4].text(-4.8, -0.3, 
            s='FormatStrFormatter("ab%dcd")', 
            fontsize=14)

plt.show()
```

### plt.style
https://matplotlib.org/stable/api/style_api.html#matplotlib-style

`plt.style.use`함수를 사용하여 `matplotlib`의 Theme을 설정할 수도 있습니다.

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 30)
y = lambda x: np.sin(x)

for style in ["dark_background",
              "classic", 
              "seaborn", 
              "ggplot"]:
    plt.style.use(style)
    fig, ax = plt.subplots(
        figsize=(6,1.5),
        subplot_kw={
        "yticks": [],
        "xticks": []
    })
    ax.set_title(style, 
                 fontsize=14)
    
    ax.plot(x, y(x))
    ax.plot(x, y(x+1))
    ax.plot(x, y(x+2))
    ax.plot(x, y(x+3))
    ax.plot(x, y(x+4))

plt.show()
```

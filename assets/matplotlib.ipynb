### plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib-axes-axes-plot

`plot` 메서드를 통해 라인플롯을 그릴 수  있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(
    figsize=(4.5, 3.0)
)
x = np.linspace(-10, 10, 1000)
y = lambda x: np.sin(x)
ax.plot(x, y(x))
ax.grid()
plt.show()
```

### scatter
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib-axes-axes-hist

`scatter` 메서드를 통해 스캐터플롯을 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(
    figsize=(4.5, 3.0)
)
x = np.linspace(-10, 10, 1000)
y = lambda x: np.sin(x)
noise = np.random.normal(
    loc=0, 
    scale=0.1, 
    size=x.shape
)

ax.scatter(x=x, 
           y=y(x)+noise, 
           s=2.5, 
           color="magenta")
ax.grid()
plt.plot()
```

### bar
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib-axes-axes-bar

`bar` 메서드를 통해 막대그래프를 그릴 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(
    figsize=(4.5, 3.0)
)
x = np.arange(-10, 10)
y = lambda x: np.sin(x)
noise = np.random.normal(
    loc=0, 
    scale=0.1, 
    size=x.shape
)

ax.bar(x=x, 
       height=y(x)+noise, 
       color="darkblue")
ax.grid()
plt.plot()
```

### ticker.Locator
https://matplotlib.org/stable/api/ticker_api.html#tick-locating

`plt.ticker`에는 $x$ 혹은 $y$축의 눈금(tick)을 설정할 수 있는 Locator 클래스들이 있습니다. 

```python
from matplotlib import ticker
import matplotlib.pyplot as plt
fig, axes = plt.subplots(
    nrows=4,
    figsize=(4.5,7),
    subplot_kw={
        "xlim": [-5, 5],
        "ylim": [-0.5, 0.5],
        "yticks": [],
    },
)

for ax in axes:
    ax.spines[["top", "left", "right"]].set_visible(False)


axes[0].xaxis.set_major_locator(
    ticker.NullLocator()
)
axes[0].text(-4.8, -0.3, 
            s= "NullLocator()", 
            fontsize=14)


axes[1].xaxis.set_major_locator(
    ticker.FixedLocator([-5, 0, 5])
)
axes[1].text(-4.8, -0.3, 
        s='FixedLocator([-5, 0, 5])', 
        fontsize=14)

axes[2].xaxis.set_major_locator(
    ticker.MultipleLocator(2.5)
)
axes[2].text(-4.8, -0.3, 
            s='ticker.MultipleLocator(2.5)', 
            fontsize=14)


axes[3].xaxis.set_major_locator(
    ticker.FixedLocator([-5, 0, 5])
)
axes[3].xaxis.set_minor_locator(
    ticker.MultipleLocator(1)
)
axes[3].text(-4.8, -0.3, 
    s="""set_minor_locator(
    ticker.MultipleLocator(1))""", 
    fontsize=14)

plt.show()
```

### ticker.Formatter
https://matplotlib.org/stable/api/ticker_api.html#tick-formatting

Locator와 더불어 Formatter 클래스들을 통해 눈금에 대한 레이블을 지정할 수 있습니다.

```python
from matplotlib import ticker
import matplotlib.pyplot as plt
fig, axes = plt.subplots(
    nrows=5,
    figsize=(4.5,10),
    subplot_kw={
        "xlim": [-5, 5],
        "ylim": [-0.5, 0.5],
        "yticks": [],
    },
)

for ax in axes:
    ax.spines[[
        "top", 
        "left", 
        "right"]
    ].set_visible(False)
    
    ax.xaxis\
    .set_major_locator(
        ticker.FixedLocator([-5, 0, 5])
    )

axes[0].xaxis.set_major_formatter(
    ticker.NullFormatter()
)
axes[0].text(-4.8, -0.3, 
            s= "NullFormatter()", 
            fontsize=12)


axes[1].xaxis.set_major_formatter(
    ticker.FixedFormatter(
        ["a","b","c"])
)
axes[1].text(
    -4.8, 
    -0.3, 
    s='FixedFormatter(["a","b","c"])', 
    fontsize=12)


axes[2].xaxis.set_major_formatter(
    ticker.FuncFormatter(
        lambda x, pos: f"{x}xyz"),
)
axes[2].text(
    -4.8, 
    -0.3, 
    s='''FuncFormatter(
    lambda x, pos: f"{x}xyz")''', 
    fontsize=12
)


axes[3].xaxis.set_major_formatter(
    ticker.PercentFormatter(xmax=5)
)
axes[3].text(-4.8, -0.3, 
    s="PercentFormatter(xmax=5)", 
    fontsize=12)


axes[4].xaxis.set_major_formatter(
    ticker.FormatStrFormatter("ab%dcd")
)
axes[4].text(-4.8, -0.3, 
    s='FormatStrFormatter("ab%dcd")', 
    fontsize=12)

plt.show()
```

### plt.subplots_adjust
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html#matplotlib-pyplot-subplots-adjust


`plt.subplots_adjust` 함수를 통해 `subplot`들을 조정할 수 있습니다.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(
    nrows=2, 
    ncols=2,
    figsize=(4, 2))
fig.suptitle(
    "# not adjusted",
    fontsize=12, 
    ha="left",
    x=0.1,
    va="bottom",
    color="gray")
plt.show()

fig, ax = plt.subplots(
    nrows=2, 
    ncols=2,
    figsize=(4, 2))
plt.subplots_adjust(hspace=1.5)
fig.suptitle(
    "plt.subplots_adjust(vspace=1.5)",
    fontsize=12, 
    ha="left",
    x=0.1,
    va="bottom")
plt.show()


fig, ax = plt.subplots(
    nrows=2, 
    ncols=2,
    figsize=(4, 2))
plt.subplots_adjust(wspace=1.5)
fig.suptitle(
    "plt.subplots_adjust(wspace=1.5)",
    fontsize=12, 
    ha="left",
    x=0.1,
    va="bottom")

plt.show()

```

### Text Align
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html#matplotlib-pyplot-text



```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(
    figsize=(4.5,3),
    subplot_kw={
        "xticks": np.linspace(0, 1, 11),
        "yticks": np.linspace(0, 1, 11)
    }
)
ax.grid()
ax.text(
    0.5, 
    0.5, 
    "bottom-right", 
    fontsize=15,
    alpha=0.5,
    color="blue",
    va="bottom",
    ha="right")

ax.text(
    0.5, 
    0.5, 
    "top-left", 
    fontsize=15,
    alpha=0.5,
    color="red",
    ha="left",
    va="top")
ax.text(
    0.5, 
    0.5, 
    "center-center", 
    fontsize=15,
    alpha=0.5,
    color="black",
    ha="center",
    va="center")


plt.show()
```

### Legend placement
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html#matplotlib-axes-axes-legend



```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(
    nrows=4,
    ncols=1,
    figsize=(4,8),
    subplot_kw={
        "xlim": [0, 1],
        "ylim": [0, 1]
    }

)
for ax, loc in zip(axes, [
    "upper right",
    "center left",
    "center",
    "lower center",
]):
    ax.scatter(
        np.random.normal(
            loc=0.5,
            scale=0.5,
            size=8
        ), 
        np.random.normal(
            loc=0.5,
            scale=0.5,
            size=8
        )
    )
    ax.legend(labels=[loc], loc=loc)
plt.show()
```

### Interpolation for imshow
https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html#interpolations-for-imshow

해당 코드는 [Matplotlib 공식문서](https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html#interpolations-for-imshow)에서 가져왔습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

methods = [
    None, 
    'none', 
    'nearest', 
    'bilinear', 
    'bicubic', 
    'spline16',
    'spline36', 
    'hanning', 
    'hamming', 
    'hermite', 
    'kaiser', 
    'quadric',
    'catrom', 
    'gaussian', 
    'bessel', 
    'mitchell', 
    'sinc', 
    'lanczos'
]

# Fixing random state for reproducibility
np.random.seed(19680801)

grid = np.random.rand(4, 4)

fig, axs = plt.subplots(nrows=6, ncols=3, figsize=(4.5, 6),
                        subplot_kw={'xticks': [], 'yticks': []})

for ax, interp_method in zip(axs.flat, methods):
    ax.imshow(grid, interpolation=interp_method, cmap='viridis')
    ax.set_title(str(interp_method))

plt.tight_layout()
plt.show()
```

### Colormaps
https://matplotlib.org/stable/tutorials/colors/colormaps.html#choosing-colormaps-in-matplotlib

`plt.colormaps()`를 호출하면 `matplotlib`안에 담긴 다양한 `colormap`들을 확인할 수 있습니다.

```python
import matplotlib.pyplot as plt 
import re 
import numpy as np
from IPython.display import (
    display, 
    display_html, 
    HTML)

def display_colormap(cmap: str):
    html_text = (
        plt.colormaps[cmap]
        ._repr_html_()
    )
    strong_elem = re.search(
        r"(?<=\<strong\>).*(?=\</strong\>)", 
        html_text)[0]

    base64img = re.search(
        r'data:image/png;.+?(?=">)', 
        html_text)[0]


    display_html(HTML(
        f'''<p>
<code>plt.colormaps[{strong_elem}]</code></p>
<p><img 
src="{base64img}" 
style="width: 240px; height: 5px;">
</p>'''
    ))

for cmap in plt.colormaps()[:5]:
    display_colormap(cmap)
    
```

### plt.style
https://matplotlib.org/stable/api/style_api.html#matplotlib-style

`plt.style.use`함수를 사용하여 `matplotlib`의 Theme을 설정할 수도 있습니다.

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 30)
y = lambda x: np.sin(x)

for style in ["dark_background",
              "classic", 
              "seaborn", 
              "ggplot"]:
    plt.style.use(style)
    fig, ax = plt.subplots(
        figsize=(4.5, 1.5),
        subplot_kw={
        "yticks": [],
        "xticks": []
    })
    ax.set_title(style, 
                 fontsize=14)
    
    ax.plot(x, y(x))
    ax.plot(x, y(x+1))
    ax.plot(x, y(x+2))
    ax.plot(x, y(x+3))
    ax.plot(x, y(x+4))

plt.show()
```
